
############################
# 一、服务管理（Linux）
############################

sudo systemctl start mysql        # 启动 MySQL
sudo systemctl stop mysql         # 停止 MySQL
sudo systemctl restart mysql      # 重启 MySQL
sudo systemctl status mysql       # 查看状态
sudo systemctl enable mysql       # 开机自启
sudo systemctl disable mysql      # 取消开机自启

# 查看进程与端口
ps -ef | grep mysql
netstat -tlnp | grep 3306
ss -tlnp | grep 3306


############################
# 二、登录 MySQL
############################

mysql -u root -p                  # 本地登录
mysql -h 127.0.0.1 -P 3306 -u root -p   # 指定端口
mysql -h <ip地址> -u root -p      # 远程登录

mysql -u interview_agent -p      # 登录: 
SELECT CURRENT_USER();           # 确认现在是谁登录的

############################
# 三、数据库操作（SQL）
############################

SHOW DATABASES;                   # 查看数据库
CREATE DATABASE test_db;          # 创建数据库
DROP DATABASE test_db;            # 删除数据库
USE test_db;                      # 进入数据库


############################
# 四、表操作
############################

SHOW TABLES;                      # 查看表
CREATE TABLE user (
  id INT PRIMARY KEY AUTO_INCREMENT,
  name VARCHAR(50),
  age INT
);

DESC user;                        # 表结构
DROP TABLE user;                  # 删除表


############################
# 五、CRUD 操作
############################

INSERT INTO user(name, age) VALUES ('Tom', 20);   # 插入
SELECT * FROM user;                                # 查询
UPDATE user SET age = 25 WHERE name='Tom';         # 更新
DELETE FROM user WHERE name='Tom';                 # 删除


############################
# 六、用户与权限
############################

CREATE USER 'test'@'%' IDENTIFIED BY '123456';     # 创建用户

GRANT ALL PRIVILEGES ON test_db.* TO 'test'@'%';   # 授权

FLUSH PRIVILEGES;                                  # 刷新权限

SHOW GRANTS FOR 'test'@'%';                        # 查看权限

DROP USER 'test'@'%';                              # 删除用户


############################
# 七、备份与恢复
############################

# 备份数据库
mysqldump -u root -p test_db > test_db.sql

# 备份全部数据库
mysqldump -u root -p --all-databases > all.sql

# 恢复数据库
mysql -u root -p test_db < test_db.sql
